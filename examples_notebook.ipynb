```python
from openai import OpenAI
from dotenv import load_dotenv
import os
from src.cleanlab_tlm.utils.responses import TLMResponses

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
validator = TLMResponses(
    api_key=os.getenv("CLEANLAB_TLM_API_KEY"),
    options={
      "log": ["explanation"]
    }
)
openai_kwargs = {"model": "gpt-4.1-mini", "temperature": 0.0}
```

## Single turn QA


### Example 1


```python
query = """I want to lease a car, calculate my monthly payment. Here are the details:

car price: 38k
sales tax: 8%s
down payment: 5k
lease term: 36 months
money factor: 0.0025
residual value: 25k

"""
response = client.responses.create(input=query, **openai_kwargs)
print(response.output_text)
```

```python
tlm_response = validator.score(
    input=query, response=response, **openai_kwargs
)
print(tlm_response["trustworthiness_score"])
print(tlm_response["log"]["explanation"]) # type: ignore
```

## File Search


```python
# Create OpenAI Knowledge Base containing this file

file = client.files.create(file=open("openai-pricing.pdf", "rb"), purpose="user_data")

vector_store = client.vector_stores.create(name="knowledge_base")

client.vector_stores.files.create_and_poll(vector_store_id=vector_store.id, file_id=file.id)

print("Created vector store with ID:", vector_store.id)

openai_kwargs["store"] = False
rag_kwargs = {
    "tools": [{"type": "file_search", "vector_store_ids": [vector_store.id]}],
    "include": ["file_search_call.results"],
    # "instructions": "Do NOT use your own knowledge, only use the file search tool to answer the question",
    **openai_kwargs,
}
```

```python
query = "How much gpt-5 cost?"


response = client.responses.create(input=query, **rag_kwargs)
print(response.output_text)
```
